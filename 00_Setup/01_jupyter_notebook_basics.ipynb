## Variables in Python 

Variables are containers for storing data values.

- In Python, variables are created when you assign a value to it.
- Python has no command for declaring a variable.

```python
name = "Sree"
age = 25

# Here, 'name' and 'age' are variables which are assigned the values "Sree" and '25'.
# We can now use 'name' and 'age' to access the respective values

print("name = ",name)
print("age = ",age)
```

Variables do not need to be declared with any particular type, and can even change type after they have been set.

```python
name = 32
age = "Twenty Five"

# Changed the value and type of variables 'name' and 'age'

print("name = ",name)
print("age = ",age)
```

## Arithmetic Operations

We use various operators for doing arithmetic operations. Some basic arithmetic operators are:
- '+' to add two variables (for numeric values)
- '-' to subtract two variables
- '*' to multiply two variables
- '/' to divide two variables

Here we are seeing and executing these basic operators. There are more operators such as Assignment operators, Comparison operators, Logical Operators etc., which we will study later.

```python
## Arithmetic Operations

a = 10
b = 5

# Here, 'a' is a variable with value '10' and 'b' is a variable with value '5'

# Basic Arithmetic Operations

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```
